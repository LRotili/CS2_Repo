## 04) deg

```python
edges = []

with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/deg.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        for line in file:
            u, v = map(int, line.split())
            edges.append((u, v))


def deg(n, edges):
    D = [0] * (n + 1)    
    for u, v in edges:
        D[u] += 1 
        D[v] += 1  
    return D[1:]

degrees = deg(n, edges)
print(" ".join(map(str, degrees)))
```

## 05)ddeg

```python
edges = []
with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/ddeg.txt', 'r') as file:
    n, m = map(int, file.readline().split()) 
    for line in file:
        u, v = map(int, line.split())
        edges.append((u, v))  

def ddeg(n, edges):
    degree = [0] * (n + 1)  
    neighbors = [[] for _ in range(n + 1)] 
    
    for u, v in edges:
        degree[u] += 1  
        degree[v] += 1  
        neighbors[u].append(v)  
        neighbors[v].append(u)      
    sum_of_degrees = []
    for i in range(1, n + 1):
        sum_neighbors_degrees = sum(degree[neighbor] for neighbor in neighbors[i])
        sum_of_degrees.append(sum_neighbors_degrees)  
    return sum_of_degrees

result = ddeg(n, edges)
print(" ".join(map(str, result)))
```

## 06)cc

```python
edges = []
with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/cc.txt', 'r') as file:
    n, m = map(int, file.readline().split()) 
    for line in file:
        u, v = map(int, line.split())
        edges.append((u, v))  

def dfs(graph, visited, node):
    visited[node] = True
    for neighbor in graph[node]:
        if not visited[neighbor]:
            dfs(graph, visited, neighbor)

def cc(n, edges):
    graph = [[] for _ in range(n + 1)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    visited = [False] * (n + 1)
    component_count = 0
    for node in range(1, n + 1):
        if not visited[node]:
            dfs(graph, visited, node)
            component_count += 1
    return component_count

result = cc(n, edges)
print(result)
```

## 07)bfs

```python
edges = []
with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/bfs.txt', 'r') as file:
    n, m = map(int, file.readline().split()) 
    for line in file:
        u, v = map(int, line.split())
        edges.append((u, v))  

def bfs(n, edges):
    graph = [[] for _ in range(n + 1)]  
    for u, v in edges:
        graph[u].append(v)
    D = [-1] * (n + 1)
    D[1] = 0
    queue = [1]
    while queue:
        current = queue.pop(0)  
        for neighbor in graph[current]:
            if D[neighbor] == -1: 
                D[neighbor] = D[current] + 1 
                queue.append(neighbor)  
    return D[1:]

result = bfs(n, edges)
print(" ".join(map(str, result)))
```

## 08) dij

```python
import networkx as nx

with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/dij.txt', 'r') as file:
    G = nx.DiGraph() 
    first_line = file.readline()
    num_nodes, num_edges = map (int, first_line.split())

    lines = file.readlines()
    
    for line in lines:
            node1, node2, weight = map(int, line.split())
            G.add_edge(node1, node2, weight=weight)
    
def find_shortest_paths(G, num_nodes):
    D = [-1] * num_nodes
    D[0] = 0 
    for i in range(1, num_nodes):
        try:
            D[i] = nx.dijkstra_path_length(G, 1, i + 1) 
        except nx.NetworkXNoPath:
            D[i] = -1 
    return D

D = find_shortest_paths(G, num_nodes)

print(" ".join(map(str, D)))
```

## 09) dag 

```python
import networkx as nx

with open(r'/home/elleu/Scrivania/CS2_Repo/assignment #4/dag.txt', 'r') as file:
    lines = file.readlines() 
    graphs = []
    k = int(lines[0].strip())
    i = 1 
    for whatever in range(k):
        while i < len(lines) and lines[i].strip() == "":
            i += 1
        n, m = map(int, lines[i].strip().split())
        i += 1
        G = nx.DiGraph() 
        for whateverelse in range(m):
            u, v = map(int, lines[i].strip().split())
            G.add_edge(u-1, v-1)
            i += 1
        graphs.append(G)
        while i < len(lines) and lines[i].strip() == "":
            i += 1

def acyclic(graph):
    if nx.is_directed_acyclic_graph(graph):
        return 1 
    else:
        return -1 

results = [str(acyclic(graph)) for graph in graphs]
print(" ".join(results))
```
